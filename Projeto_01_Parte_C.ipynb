## Problema de negócio - Precificação de veículos usados

Assim, saber quanto vale um veículo não é uma questão trivial, e sim uma questão que influencia nas nossas tomadas de decisão (ENGERS; HARTMANN; STERN, 2009; OU et al., 2020).

Dessa forma, a criação de um ***modelo preditivo***, também conhecido como ***Modelo de Machine Learning***, que balanceie adequadamente o preço de veículos usados e seminovos, para que nem o comprador e nem o vendedor sejam prejudicados, é assim necessário uma vez que possibilita que tomemos decisões baseadas em dados. ### Métodos para a construção, métrica e método de validação do modelo de Machine Learning

<img src="https://raw.githubusercontent.com/Campos-Silva/Projeto_01_Parte_C_Modelos_de_Machine_Learning_no_Python/main/metodo_b.jpg" width="400" align="left" />

Para demonstrar como realizamos a seleção de modelos através da linguagem python eu realizei essa ação através de **duas etapas**, denominadas como etapas "1" e "2".

A seguir descrevo como as realizei:

### Etapa 1

- Nessa etapa, demostrei como realizar a seleção de um modelo de machine learning de forma manual, indo passo a passo através de seleção por meio de três tipos de algoritimos. Os algoritimos de Machine Learning que selecionei são: **Regressão Linear Simples**, **Vizinhos mais próximos** e **Árvores de Decisão**.


- Para a escolha do Modelo final utilizei como métrica o "**Erro Quadrático Médio**"'. O EQM toma como base a diferença entre os valores da predição do modelo com os valores reais  já conhecidos no conjunto de dados, sendo esse o valor de erro do modelo. Para que não seja gerado valores negativos, esses valores são elevados ao quadrado, e posteriormente é gerado um valor médio do erro.


- A fim de auxiliar também na escolha do modelo final utilizei um método de validação. O método de validação tem o objetivo de acessar a qualidade da métrica utilizada, que nesse caso é o EQM. O método que utilizei foi o **Método de Holdout Repetido**. Esse método envolve a divisão aleatória do conjunto de dados em: (1) **treinamento**, o qual servirá para treinar o modelo e (2) **teste**, o qual será utilizado para avaliar o desempenho do modelo treinado (UFPR, 2021). Para esse projeto utilizei 100 repetições do método de Holdout.


- O **modelo final escolhido** foi aquele que apresentou **o menor valor de EQM dentre esses tipos de modelos** utilizando o método de Holdout Repetido. Para essa etapa utilizei apenas duas variáveis que são a kilometragem e o consumo de combustível por litro.

### **Etapa 2**

Nessa etapa utilizei a automatização de seleção de modelos através da biblioteca [Pycaret](https://pycaret.org/). É uma biblioteca que facilita a implementação de seleção de modelos através de argumentos com poucas linhas de código. Para a seleção do modelo final segui as seguintes etapas para a construção do modelo:

- Fiz a seleção de modelos utilizando todos os algoritimos de machine learning disponibilizados pela biblioteca através da da função "pycaret.regression import * ".


- Para a etapa de construção do modelo utilizei todas as oito variáveis dos veículos.


- Utilizei os seguintes hiperparâmetros na função "setup": "normalize = True", "n_jobs = 2", "ignore_low_variance = True","remove_multicollinearity = True" e "multicollinearity_threshold = 0.95".


- Logo após iss fiz a seleçao dos 5 tipos de modelos com os menores valores de RMLSE através da função "compare_models".


- A métrica de erro de "RMLSE" penaliza mais o erro para baixo do valor real do que para cima do valor real. Dessa forma, é uma métrica que se adequa ao setor de negócios em questão. Isso porque se o modelo prever um valor abaixo do valor real, o vendedor estará perdendo dinheiro na venda desse veículo. Por exemplo, se o modelo *prever que o preço de venda de um veículo for de 5 mil dólares, quando na verdade é 7 mil dólares, o vendedor estará perdendo 2 mil dólares*. Assim, optei por escolher o **modelo final com menor valor de RMSLE**, uma vez que será escolhido o modelo preditivo que tende a errar menos para baixo do valor real.


- Os cinco tipos de modelos com menor valor de RMLSE foram incrementados com hiperparâmetros através da função "tune_model".


- Para finalizar a seleção do modelo final foi utilizado a função "blend_models", o qual une os modelos selecionados junto com os seus hiperparâmetros.

# Etapa 1

Vou começar a fazer a seleção de modelos de Machine Learning sem automatização via pycaret. Farei isso para demonstrar como é feito a seleção de modelos através de três tipos de algoritimos de Machine Learning.

Para isso vou fazer apenas com duas variáveis do Dataset, que serão a kilometragem e o tipo de combustível.

A fim de deixar as variáveis analisadas com escala aproximada vou dividir o preço e o consumo do combustível.

```python
# Dividindo as variáveis por mil e criando novas colunas com essas variáveis recem criadas

dados['preco_de_venda'] = dados['preco_de_venda_USD']/1000

dados['km_rodado_d_1000'] = dados['kilometragem']/1000

dados = dados[['preco_de_venda', 'km_rodado_d_1000', 'consumo_do_combustivel_kmpl']]

dados.head()
```

Agora vou fazer uma pré-visualização da relação entre preço de venda e kilometragem para verificar se há uma possível relação de tendência entre elas.
Para isso irei utilizar um gráfico de scatterplot


```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(x = 'km_rodado_d_1000', y = 'preco_de_venda', data = dados);

plt.title('Scatterplot para preço e kilometragem')
plt.xlabel('Kilometragem (em unidades de milhares)')
plt.ylabel('Preço em dólares (em unidades de milhares)')
```

## Regressão Linear

Agora irei construir o primeiro modelo. Ele será o modelo de regressão linear.

```python
from sklearn.linear_model import LinearRegression

# Modelo para prever a variável preço a partir da kilometragem

X = dados[['km_rodado_d_1000']]
y = dados['preco_de_venda'] 

regressao = LinearRegression()
regressao.fit(X, y) #treinamento do modelo
```

Nesse cenário, o modelo de regressão linear é dado pela fórmula a seguir:

preco = b0 + b1 * kilometragem

Agora irei identificar os coeficientes ajustados desse modelo

```python
# Resgatando os coeficientes ajustados

print('b0 = ', regressao.intercept_)
print('b1 = ', regressao.coef_ [0] )
```

A fórmula para esse modelo agora pode ser definida e é dada da seguinte maneira:

Preço = 8.23 - 0.033 * kilometragem

Com o modelo criado a partir da função "regressao" é possível agora prever qual o preço a partir da entrada de valores de kilometragem. Isso será feito abaixo.

```python
novos_dados = pd.DataFrame({'km_rodado_d_1000': [27, 95, 112, 165]})
regressao.predict(novos_dados)
```

Agora irei fazer um gráfico contendo a curva ajustada para esse modelo 1.

```python
# Gráfico do Ajuste

b0 = regressao.intercept_
b1 = regressao.coef_

import numpy as np
pontos_em_x = np.arange(0, 200, 0.1)
pontos_em_y = b0 + b1 * pontos_em_x

sns.scatterplot(x = 'km_rodado_d_1000', y = 'preco_de_venda', data = dados)

import matplotlib.pyplot as plt
plt.plot(pontos_em_x, pontos_em_y, color = 'blue')
plt.title('Ajuste do Modelo  1 da Regressão Linear')
plt.xlabel('Kilometragem (em unidades de milhares)')
plt.ylabel('Preço em dólares (em unidades de milhares)')
plt.show()
```

### **Métrica para Avaliação do Modelo**

Agora irei fazer um avaliação desse modelo utilizando a métrica de Erro Quadrático Médio (EQM).

avaliação.jpg

```python
# Fórmula para calcular o EQM do Modelo 1

preditos = regressao.predict(X)
eqm = ((y - preditos)**2).mean()
eqm
```

```python
# Função do "sklearn" para o cálculo automático do EQM para o Modelo 1

from sklearn.metrics import mean_squared_error 
eqm = mean_squared_error(y, preditos)

print('EQM da Regressão Linear do Modelo 1: {}'.format(round(eqm, 2)))
```

### **Estratégia de Validação do Modelo**

Agora irei utiizar o método de Holdout para validar essa métrica do modelo 1

```python
# Importando a função de holdout do sklearn

from sklearn.model_selection import train_test_split 
```

```python
#Realizando a divisão do conjunto de dados em dados de treino e teste.

#Aqui irei utilizar 25% do conjunto de dados para teste

train_test_split(X, y, test_size = 0.25)
```

Agora irei calcular o EQM médio obtido através desse método de validação Holdout

```python
# Primeiro irei separar os dados em teste e treinamento 
Xtreino, Xteste, ytreino, yteste = train_test_split(X, y, test_size = 0.25)

# Agora irei ajustar o modelo aos dados de treino
regressao = LinearRegression()
regressao.fit(Xtreino, ytreino)

# Posteriormente utilizarei o modelo ajustado para predizer os dados de teste
preditos = regressao.predict(Xteste) 

# Agora irei calcular o EQM nos dados de teste 
eqm = mean_squared_error(yteste, preditos)
print('EQM : {}'.format(round(eqm, 2)))
```

Observação: o método holdout gera distintos valores de EQM  a medida que é atualizado essas linhas de códigos acima. Isso ocorre porque são pegos aleatoriamente novos dados para treino e teste a medida que a técnica é realizada. Dessa forma, o resultado por si só gera variabilidade a medida que ocorre essas atualizações. Isso gera incerteza  na estimativa de EQM.

Uma forma de diminuir essa variabilidade e gerar valores de EQM com menor grau de incerteza no valor final de EQM é realizar repetições do método Holdout e fazer a média do EQM gerado pelo número total de repetições. Essa técnica é chamada de **Holdout Repetido**. Isso será feita logo mais. Porém, irei realizar antes o modelo de regressão linear agora com duas variáveis para checar se há uma diminuição do valor de EQM.

```python
# Modelo 2 - Realizando o modelo de regressão linear agora com duas variáveis preditoras

# Criando as variáveris preditoras ("X") e resposta ("y")

X = dados[['km_rodado_d_1000', 'consumo_do_combustivel_kmpl']]
y = dados['preco_de_venda'] 
```

Nesse cenário, o modelo 2 de regressão linear é dado por

 **Preço = b0 + b1 * kilometragem + b2 * consumo de combustivel**

Agora irei checar o valor de EQM através do método Holdout para esse modelo 2

```python
# Realizando o método holdout para o modelo 2


Xtreino, Xteste, ytreino, yteste  = train_test_split(X, y, test_size = 0.25)
regressao = LinearRegression()
regressao.fit(Xtreino, ytreino)
preditos = regressao.predict(Xteste)
eqm = mean_squared_error(yteste, preditos)

print('EQM para o Modelo 2: {}'.format(round(eqm, 2)))
```

Com o resultado acima já é possivel inferir que esse modelo 2 possui melhor eficiencia uma vez que ele possui menor valor de EQM. Ou seja, há menor erro na estimativa final do preço. Assim, a variável consumo de combustível anexada ao modelo ajuda na melhora da eficiencia do modelo.

Agora irei checar quais são os coeficientes da fórmula para esse modelo 2. 

```python
# Coeficientes
b0 = regressao.intercept_
b1, b2 = regressao.coef_
print('b0 = ', b0)
print('b1 = ', b1)
print('b2 = ', b2)
```

Dessa forma, a fórmula do modelo 2 é dada da seguinte forma:

Preço = 1.00 - 0.033 * kilometragem + 0.34 * consumo de combustível

```python
# Gráfico do Ajuste com DUAS Variáveis
sns.scatterplot(x = 'km_rodado_d_1000', y = 'preco_de_venda', hue = 'consumo_do_combustivel_kmpl', data = dados)
plt.title('Ajuste do Modelo 2 de Regressão Linear')
plt.xlabel('Kilometragem (em unidades de milhares)')
plt.ylabel('Preço em dólares (em unidades de milhares)')
plt.show()
```

Agora irei realizar o Método de Holdout Repetido para os modelos 1 e 2 para checar se há diminuição da variabilidade no valor de EQM.

### Método de Validação de Holdout Repetido

Modelo 1 - Uilizando uma variável

```python
# Colocando o modelo em produção de forma básica

def PrecoVeiculo(kilometragem, tipo_combustivel, tipo_do_vendedor, transmissao, dono, consumo_do_combustivel_kmpl, motor_CC, potencia_do_motor_bhp, Marca):
    dados0 = {'kilometragem': [kilometragem], 'tipo_combustivel': [tipo_combustivel], 'tipo_do_vendedor': [tipo_do_vendedor], 'transmissao': [transmissao], 'dono': [dono], 'consumo_do_combustivel_kmpl': [consumo_do_combustivel_kmpl], 'motor_CC': [motor_CC], 'potencia_do_motor_bhp': [potencia_do_motor_bhp], 'Marca': [Marca]}
    dados = pd.DataFrame(dados0)

    modelo = load_model('meu-melhor-modelo-para-precos-veiculos')

    pred = float(predict_model(modelo, data = dados)['Label'].round(2))
    valor = round(1.8*pred, 2)
    
    print(' ')
    print('Kilometragem: {}'.format(kilometragem))
    print('Tipo do Combustível: {}'.format(tipo_combustivel))
    print('Tipo do Vendedor: {}'.format(tipo_do_vendedor))
    print('Tipo de Transmissão: {}'.format(transmissao))
    print('Tipo do dono: {}'.format(dono))
    print('Consumo do Combustível (Km/L): {}'.format(consumo_do_combustivel_kmpl))
    print('Cilindradas do Motor (CC): {}'.format(motor_CC))
    print('Potência do Motor (BHP): {}'.format(potencia_do_motor_bhp))
    print('Marca: {}'.format(Marca))
    print('Preço estimado do Veículo: ${}'.format(pred))
    print('Valor de Venda: ${}'.format(valor))
```
